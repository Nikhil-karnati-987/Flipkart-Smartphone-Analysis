```python
df['RAM']=df['RAM'].str.replace("GB","").astype(int)
df['ROM']=df['ROM'].str.replace("GB","").astype(int)
df['Display_size']=df['Display_size'].str.replace("inch","").astype(float)
df['Camera']=df['Camera'].str.replace("MP","").astype(int)
df['Original_price']=df['Original_price'].str.replace(",","").astype(int)
df['offer_price']=df['offer_price'].str.replace(",","").astype(int)
```

```python
# Assuming 'df' is your DataFrame and 'rating' is the column you want to check
df = df[df['Rating'] <= 5.0]  # This keeps only rows where the rating is less than or equal to 5

# Optionally, reset the index if needed (to clean up row indices after the rows are dropped)
df.reset_index(drop=True, inplace=True)
```

```python
df.info()
```

```python
# Assuming your DataFrame has 'Original Price' and 'Offer Price' columns
df['Discount (%)'] = round(((df['Original_price'] - df['offer_price']) / df['Original_price']) * 100,1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average offer price by product
average_offer_price_by_product = df1.groupby('Brand')['offer_price'].mean().reset_index()
average_offer_price_by_product.columns = ['Brand', 'Average Offer Price']

# Sort by Average Offer Price in descending order
average_offer_price_by_product = average_offer_price_by_product.sort_values(by='Average Offer Price', ascending=False).head(10)

# Create a bar plot
plt.figure(figsize=(12, 6))
bar_plot = sns.barplot(x='Average Offer Price', y='Brand', data=average_offer_price_by_product, palette='viridis')

# Add annotations for each bar
for index, value in enumerate(average_offer_price_by_product['Average Offer Price']):
    bar_plot.text(value, index, f'{value:.2f}', color='black', ha='left', va='center')

plt.title('Average Offer Price by Product')
plt.xlabel('Average Offer Price')
plt.ylabel('Brand')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))

# Plotting Actual Price
plt.subplot(1, 2, 1)
plt.scatter(df1['Rating'], df1['Original_price'], color='blue', alpha=0.6, label='Original_price')
plt.title('Rating vs. Original_price')
plt.xlabel('Rating')
plt.ylabel('Original_price ()')
plt.grid()
plt.legend()

# Plotting Discounted Price
plt.subplot(1, 2, 2)
plt.scatter(df1['Rating'], df1['offer_price'], color='orange', alpha=0.6, label='offer_price')
plt.title('Rating vs. offer_price')
plt.xlabel('Rating')
plt.ylabel('offer_price ()')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

Insight: The analysis shows a strong correlation between ratings and both Original and Offer prices, indicating that consumers are willing to pay more for higher-rated products. This consistency suggests that even with discounts, well-rated items maintain their perceived value, allowing businesses to justify higher price points for quality offerings.
